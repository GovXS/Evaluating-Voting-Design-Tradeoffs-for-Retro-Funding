```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Gini index results CSV file
gini_df = pd.read_csv('/Users/idrees/Code/govxs/outdir_csv/gini_results.csv')

# Create a boxplot for Gini index for different voting rules
plt.figure(figsize=(10, 6))
gini_df.boxplot(column=[col for col in gini_df.columns if 'round' not in col], grid=False)
plt.title('Gini Index', fontsize=14)
plt.ylabel('Gini Index')
plt.xlabel('Voting Rules')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the uploaded CSV file
file_path = '/Users/idrees/Code/govxs/data/vm_data/control_experiment_results_20240913_230831_40_145_8000000.0_5.csv'
control_results_df = pd.read_csv(file_path)

# Display the first few rows to understand the structure of the data
control_results_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Gini index results CSV file
gini_df = pd.read_csv('/Users/idrees/Code/govxs/data/simulation_results/small_round_10_10/gini_results.csv')

# Create a boxplot for Gini index for different voting rules
plt.figure(figsize=(10, 6))
gini_df.boxplot(column=[col for col in gini_df.columns if 'round' not in col], grid=False)
plt.title('Gini Index num_voters: small_round_10_10', fontsize=14)
plt.ylabel('Gini Index')
plt.xlabel('Voting Rules')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
